```python
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://etlReaderAnalysis:etl_reader_analysis__Gr2rEVBXyPWzIrP@34.126.84.83:27017,34.142.204.61:27017,34.142.219.60:27017/')
db = client["blockchain_etl"]  
collection = db["transactions"] 


def filter_transaction(aggregated_list, chain):
    # Set the batch size
    if chain == "bnb":
        db = client["blockchain_etl"]  
    elif chain == "eth":
        db = client["ethereum_blockchain_etl"]
    else:
        db = client["polygon_blockchain_etl"]
    collection = db["transactions"] 
    batch_size = 25

    # Initialize variables
    result = []
    i = 0

    while i < len(aggregated_list):
        start = i
        if i + batch_size < len(aggregated_list):
            end = i + batch_size
        else:
            end = len(aggregated_list)
        # Define the aggregation pipeline
        try:
            pipeline = [
                {"$match": {"to_address": {"$in": aggregated_list[start:end]}}},
                {"$group": {"_id": "$to_address", "recordCount": {"$sum": 1}}},
                {"$sort": {"recordCount": -1}}
            ]

            # Use find with batch_size and maxTimeMS
            cursor = collection.aggregate(pipeline, batchSize=batch_size, maxTimeMS=60000)

            # Extend the result with the cursor
            result.extend(list(cursor))
            print(start, end)
            print("Done {}%".format(end / len(aggregated_list) * 100))

        except:
            # Handle the timeout exception
            print(f"Timeout occurred for batch {start}-{end}, continuing to the next batch.")

        i += batch_size


    # Convert the cursor result to a DataFrame
    df = pd.DataFrame(result)

    # Print or process the DataFrame
    return df
```
