```python
with open("projects-1.json", encoding="utf-8") as f:
    kg = json.load(f)
```

```python
df = pd.DataFrame(kg).explode("deployedChains")
```

```python
df[['contract_addresses','_id','deployedChains']]
```

```python
bnb_df = df[df['deployedChains']=='0x38'].reset_index()
```

```python
bnb_df[['contract_addresses','_id','deployedChains']]
```

```python
exploded_bnb_df = bnb_df.explode('contract_addresses')
```

```python
aggregated_bnb_list = exploded_bnb_df['contract_addresses'].tolist()
aggregated_bnb_list
```

```python
file_path = 'output_bnb.txt'

# Write the list elements to the text file
with open(file_path, 'w') as file:
    file.write('[')
    for item in aggregated_bnb_list[:-1]:
        file.write('"'+ str(item) + '"'+', ')
    file.write('"'+ str(item) + '"')
    file.write(']')

print(f'Text file created at {file_path}')
```
