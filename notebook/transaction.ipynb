```python
import os
import sys
sys.path.append(os.path.dirname(sys.path[0]))

import pandas as pd
```

```python
df1 = pd.read_csv("/Users/nguyenhuyhai/20231/bigdata/group21_problem4/data/query_data/transaction/transaction_100000.csv")
df2 = pd.read_csv("/Users/nguyenhuyhai/20231/bigdata/group21_problem4/data/query_data/transaction/transaction_200000.csv")
df3 = pd.read_csv("/Users/nguyenhuyhai/20231/bigdata/group21_problem4/data/query_data/transaction/transaction_300000.csv")
df4 = pd.read_csv("/Users/nguyenhuyhai/20231/bigdata/group21_problem4/data/query_data/transaction/transaction_400000.csv")
df5 = pd.read_csv("/Users/nguyenhuyhai/20231/bigdata/group21_problem4/data/query_data/transaction/transaction_500000.csv")
df6 = pd.read_csv("/Users/nguyenhuyhai/20231/bigdata/group21_problem4/data/query_data/transaction/transaction_last.csv")
```

```python
df = pd.concat([df1,df2,df3,df4,df5,df6])
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
user_frequency = df.groupby("from_address").agg({"block_timestamp":list}).reset_index().rename({"from_address":"user","block_timestamp":"block_timestamp_list"}, axis=1)
user_frequency['id'] = '0x1_'+user_frequency['user']
```

```python
user_frequency
```

```python
user_frequency.to_csv("../data/prep_data/user_frequency_transaction.csv", index=False)
```

```python
df['value'] =df['value'].apply(eval)
```

```python
user_info = df.groupby('from_address')['value'].agg(sum).reset_index()
```

```python
user_info.rename({'from_address':'user','value':'amount'}, axis=1, inplace=True)
```

```python
user_info['id'] = '0x1_'+user_info['user']
user_info['asset'] = '0x0000000000000000000000000000000000000000'
user_info['event_type'] = 'TRANSFER_COIN'
user_info['volume_type'] = 'sending'
```

```python
user_info.to_csv("../data/prep_data/user_info_transaction.csv", index=False)
```

```python
uses_interact = df.groupby(['to_address','from_address'])['value'].count().reset_index(name='count')
```

```python
uses_interact.to_csv("../data/prep_data/uses_interact_transaction.csv", index=False)
```
